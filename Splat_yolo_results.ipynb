```python
from ultralytics import YOLO

model = YOLO('Splat_yolo/results/best_gauss.pt')

results = model.val(
    data='cub200.yaml',
    imgsz=640,
    split='val',
    project='runs/val',
    name='val_on_test_run'
)

print(results.box.map)    # Print mAP50-95
print(results.box.map50)  # Print mAP50
```
